```python
data = {
    "MSSV": [2301, 2302, 2401],
    "HoTen": ["Ly Thi Hong", "Nguyen Van Hai", "Le Phuong Anh"],
    "NamSinh": [2000, 2002, 2001],
    "DiaChi": ["HCM", "VT", "DN"]
}

df = pd.DataFrame(data)
df['GioiTinh'] = ['Nu', 'Nam', 'Nu']
print(df)

data1 = {
    "MSSV" : [2306, 2404, 2304],
    "HoTen" : ["Nguyen Van Tai", "Le Van Nam", "Phan Hong Tham"],
    "NamSinh" : [2002, 2001, 2001],
    "DiaChi" : ["NT", "VT", "HCM"],
    "GioiTinh" : ["Nam", "Nam", "Nu"]
}

df_new = pd.DataFrame(data1)
df = pd.concat([df, df_new], ignore_index=True)

print("\n", df)

df_nam = df[df["GioiTinh"] == "Nam"]
print("\nDanh sách sinh viên Nam:")
print(df_nam)


df_nu_hcm = df[(df["GioiTinh"] == "Nu") & (df["DiaChi"] == "HCM")]
print("\nDanh sách sinh viên Nu HCM:")
print(df_nu_hcm)

df_nu_hcm.to_csv("Nu_HCM.csv", index= False, encoding="utf-8")
```

```python
df = pd.read_csv('C:/Users/huynh/Documents/jupyter Running/Data Analysis/Data/sinhvien.csv', sep=';')
#df.info()
print(df.head())
#print(df)

sl_nam = (df['Gender'] == 'M').sum()
sl_nu = (df['Gender'] == 'F').sum()
print("\nsl nam: ", sl_nam)
print("\nsl nu: ", sl_nu)

#hien thi so sv nam nu ở mỗi lớp
grouped = df.groupby(['Class', 'Gender'])
print('\n SL sinh viên nam nữ của từng lớp: \n', grouped['Gender'].count())

#hien thị thông tin (theo tên alphabet và điểm cao)
grouped_mark = df.groupby('Class')
print('\n hien thị thông tin (theo tên alphabet và điểm cao)\n ', grouped_mark[['Name', 'Mark']].max())

#đổi tên
df.rename(columns= {'ID':'MSSV', 'Name':'Tên', 'BOD':'Ngày Sinh', 'Gender':'Giới Tính', 'Mark':'Điểm', 'Class':'Lớp'}, inplace=True)
print('\ndf sau khi sửa thuộc tính\n', df.head())

#DS nữ > 7
print('\n============================\nds nữ điểm hơn 7\n', df[df['Điểm'] > 7])

#DS chưa có điểm
ds_chua_co_diem = df[df['Điểm'].isna()]
ds_chua_co_diem_sorted = ds_chua_co_diem.sort_values('MSSV', ascending=True)
print('\nDS SV chưa có điểm\n', ds_chua_co_diem_sorted)
```

```python
#Thêm điểm vào NA
min_mark = df['Điểm'].min()
df['Điểm'] = df['Điểm'].fillna(min_mark)
print('\nĐiểm min: ', min_mark)
print('\nDF sau khi thêm điểm:\n', df.head(20))
```

```python
#thêm cột xếp loại
df['Xếp Loại'] = df['Điểm'].apply(lambda x : 'Giỏi' if x >= 8
                                 else ('Khá' if x >= 6 and x < 8 else 'Trung Bình'))
print('\nDF sau khi thêm Xếp Loại: \n', df.head())
```

```python
#show 2 SV Cao nhất
max_diem = df.sort_values(['Điểm'], ascending=False)
ds_2sv_max = max_diem.groupby('Lớp').head(2)
print('\nDS 2 sv cao nhất từng lớp: \n', ds_2sv_max[['Tên', 'Lớp', 'Điểm']])
```

```python
#F-> Nữ, M-> Nam
df['Giới Tính'] = df['Giới Tính'].replace({'F': 'Nữ', 'M': 'Nam'})

#lấy all sv nam
df_nam = df[df['Giới Tính'] == 'Nam']
max_diem = df_nam['Điểm'].max()
min_diem = df_nam['Điểm'].min()
mean_diem = df_nam['Điểm'].mean()

print("\nĐiểm lớn nhất của SV Nam:", max_diem)
print("\nĐiểm nhỏ nhất của SV Nam:", min_diem)
print("\nĐiểm trung bình của SV Nam:", round(mean_diem, 2)) #round() làm tròn
```
