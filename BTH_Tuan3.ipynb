```python
#Cau6
df6 = pd.read_csv('C:/Users/huynh/Documents/jupyter Running/Data Analysis/Data/student.csv', sep=';', encoding='utf-8')
print(df6.shape)

print(df6.head(10))
print('\n', df6.tail(10))
```

```python
#xem null
print(df6.isnull())
```

```python
#thuoc tinh null nhieu nhat
print(df6.isnull().sum())
print(df6.isnull().sum().idxmax())
print(df6.isnull().sum().max())
```

```python
df6['Gender'] = df6['Gender'].replace({'F' : 'Nu', 'M' : 'Nam'})
print(df6.head())
```

```python
#thay na bang diem trung binh
df6['Mark'] = pd.to_numeric(df6['Mark'], errors='coerce')
mean_mark = df6['Mark'].mean()
df6['Mark'] = df6['Mark'].fillna(mean_mark)
print(df6)
```

```python
#ktra du lieu trung lap
print(df6.duplicated())
print(df6.duplicated().sum())
```

```python
#Kiểm tra thuộc tính Mark có bị giá trị outlier. Thay đổi các giá trị >10 bằng giá trị max của thuộc tính này.
df6['Mark'] = pd.to_numeric(df6['Mark'], errors='coerce')

outlier = ((df6['Mark'] > 10) | (df6['Mark'] < 5))

print('So outlier: \n', outlier.sum())

max_mark = df6[df6['Mark'] <= 10]['Mark'].max()
print('Diem max: ', max_mark)
#Cú pháp .loc[row_condition, column_name]
df6.loc[(df6['Mark'] >= 10) | (df6['Mark'] < 5), 'Mark'] = max_mark
print(df6)
```

```python
#doi thuoc tinh
df6 = df6.rename(columns= {'ID' : 'MSSV', 'Name' : 'Ho Ten', 'BOD' : 'Ngay Sinh', 'Mark' : 'Diem', 'Class' : 'Lop'})
print(df6.head())
```
